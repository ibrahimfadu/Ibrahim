```python
import numpy as np
def outliers_treatment(data):
    Q1=np.quantile(data,0.25)
    Q3=np.quantile(data,0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    return np.clip(data,lower_bound,upper_bound)

    
```
