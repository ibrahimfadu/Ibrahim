```python
#Handale the missing value by imputing the missing values with arbitray value,mean and mode
df_copy=df.copy()
df_copy.isna().sum()
```

```python
#change age by mean
mean=df['Age'].mean()
df_copy['Age'].fillna(mean,inplace=True)
```

```python
df_copy.isna().sum()
```

```python
#change Cabin by mode
mode=df['Cabin'].mode()[0]
df_copy['Cabin'].fillna(mode,inplace=True)
df_copy.isnull().sum()
```

```python
#changing missing values with 
df_copy['Embarked'].value_counts()
df_copy['Embarked'].ffill(inplace=True)
df_copy.isnull().sum()
```

```python
#Handale the missing value by imputing the missing values using forward fill and backward fill
# Forward fill missing values
df_for=df.ffill()
# Backward fill missing values
df_bac=df.bfill()
df_bac.isna().sum()
```
